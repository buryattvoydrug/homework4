```python
import random
import numpy as np
import matplotlib.pyplot as plt
mu=1
sigma=2
gg=np.linspace(-5,5)
def exper(t):
    h=np.ones(1000, dtype=np.int16)
    v=list(map(lambda x: random.gauss(mu,sigma), h))
    k=list(filter(lambda x: x>=t,v))
    return len(k)/len(h)

# Экспериментальные значения для t=0,1,2,3
print('Экспериментальные значения для t=0,1,2,3')
print('при t=0 ',exper(0))
print('при t=1 ',exper(1))
print('при t=2 ',exper(2))
print('при t=3 ',exper(3))


bb=list(map(lambda x:exper(x),gg))
plt.plot(gg,bb,label = 'Экспериментальное решение')
def f(t):
    return 1/2*math.erfc((t-mu)/(math.sqrt(2)*sigma))
tt=np.linspace(-10,10)
pp=list(map(lambda x:f(x),tt))
plt.plot(tt,pp,label = 'Аналитическое решение')
plt.legend()
plt.title('Количество чисел больше заданного')
plt.savefig('./saved_plot2.png', dpi=300)
plt.xlabel("t")
plt.ylabel("вероятность, p")
```
