```python
import numpy as np
import matplotlib.pyplot as plt
xx=np.linspace(-6*np.pi,6*np.pi,100)
yy=np.sin(xx)/xx
plt.figure(figsize=(15, 5))
plt.plot(xx,yy)
plt.title(r'$y=\frac{sin(x)}{x}$')
ticks=np.arange(-6*np.pi,6.5*np.pi,np.pi/2)
a=np.arange(-6,6.5,0.5)
c=list(map(lambda a : str(a) +r'$\pi$',a))
plt.grid()
plt.xticks(ticks,c)
plt.savefig('./saved_plot.png', dpi=300)
```
